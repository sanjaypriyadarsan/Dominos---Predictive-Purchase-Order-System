```python
import pandas as pd

# Convert order_date and order_time into datetime objects
pizza['order_date'] = pd.to_datetime(pizza['order_date'])
pizza['order_time'] = pd.to_datetime(pizza['order_time'], format='%H:%M:%S').dt.time

# Combine date and time if needed
pizza['order_datetime'] = pd.to_datetime(pizza['order_date'].astype(str) + ' ' + pizza['order_time'].astype(str))

# Extract temporal features
pizza['day_of_week'] = pizza['order_date'].dt.day_name()
pizza['is_weekend'] = pizza['order_date'].dt.weekday >= 5
pizza['month'] = pizza['order_date'].dt.month
pizza['year'] = pizza['order_date'].dt.year
pizza['week_of_year'] = pizza['order_date'].dt.isocalendar().week
pizza['hour'] = pizza['order_datetime'].dt.hour

# Optional: Categorize time of day
def get_time_bucket(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

pizza['time_of_day'] = pizza['hour'].apply(get_time_bucket)
```

```python
from prophet import Prophet
from sklearn.metrics import mean_absolute_percentage_error
```

```python
train = daily_qty.iloc[:-28]
test = daily_qty.iloc[-28:]
```

```python
# pip install statsmodels pmdarima
from pmdarima import auto_arima
from statsmodels.tsa.arima.model import ARIMA

# --- automatic order search (quick & easy) ---
auto = auto_arima(train['y'], seasonal=False, trace=True,
                  error_action='ignore', suppress_warnings=True)
p,d,q = auto.order        # best order suggested

# --- fit final ARIMA ---
model = ARIMA(train['y'], order=(p,d,q))
fit   = model.fit()

# --- 7â€‘day forecast ---
fc = fit.get_forecast(steps=7).predicted_mean
fc.index.name = 'ds'
forecast_arima = fc.reset_index().rename(columns={0:'yhat'})

# --- evaluate on holdâ€‘out (28 days) ---
fc_val = fit.get_forecast(steps=28).predicted_mean
mape_arima = mean_absolute_percentage_error(test['y'], fc_val)
print(f"ARIMA 28â€‘day MAPE: {mape_arima:.2%}")
```

```python
# pip install statsmodels pmdarima
from pmdarima import auto_arima
from statsmodels.tsa.statespace.sarimax import SARIMAX

# --- automatic seasonal order search (weekly seasonality, m=7) ---
auto = auto_arima(train['y'], m=7, seasonal=True, trace=True,
                  error_action='ignore', suppress_warnings=True)
(p,d,q), (P,D,Q,m) = auto.order, auto.seasonal_order

# --- fit final SARIMA ---
model = SARIMAX(train['y'],
                order=(p,d,q),
                seasonal_order=(P,D,Q,m),
                enforce_stationarity=False,
                enforce_invertibility=False)
fit   = model.fit(disp=False)

# --- 7â€‘day forecast ---
fc = fit.get_forecast(steps=7).predicted_mean
fc.index.name = 'ds'
forecast_sarima = fc.reset_index().rename(columns={0:'yhat'})

# --- evaluate on holdâ€‘out ---
fc_val = fit.get_forecast(steps=28).predicted_mean
mape_sarima = mean_absolute_percentage_error(test['y'], fc_val)
print(f"SARIMA 28â€‘day MAPE: {mape_sarima:.2%}")
```

```python
# pip install tensorflow scikit-learn
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler

# --- scale data ---
scaler = MinMaxScaler()
y_scaled = scaler.fit_transform(daily_qty[['y']].values)

# --- create sequences (lookâ€‘back = 14 days) ---
seq_len = 14
X, y = [], []
for i in range(seq_len, len(y_scaled)):
    X.append(y_scaled[i-seq_len:i, 0])
    y.append(y_scaled[i, 0])
X, y = np.array(X), np.array(y)
X = X.reshape((X.shape[0], X.shape[1], 1))

# --- train/test split matching earlier 28â€‘day holdâ€‘out ---
X_train, X_test = X[:-28], X[-28:]
y_train, y_test = y[:-28], y[-28:]

# --- build & train LSTM ---
model = Sequential([
    LSTM(64, input_shape=(seq_len, 1), return_sequences=False),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=50, batch_size=16, verbose=0)

# --- forecast next 7 days iteratively ---
last_seq = y_scaled[-seq_len:].reshape(1, seq_len, 1)
lstm_preds = []
for _ in range(7):
    next_val = model.predict(last_seq, verbose=0)[0][0]
    lstm_preds.append(next_val)
    last_seq = np.append(last_seq[:,1:,:], [[[next_val]]], axis=1)

# --- inverse scale & build forecast df ---
lstm_preds_inv = scaler.inverse_transform(np.array(lstm_preds).reshape(-1,1)).flatten()
future_dates = pd.date_range(start=daily_qty.index.max() + pd.Timedelta(days=1), periods=7)
forecast_lstm = pd.DataFrame({'ds': future_dates, 'yhat': lstm_preds_inv})

# --- evaluate on 28â€‘day holdâ€‘out ---
#   (need rolling predictionsâ€”simplest quick evaluation:)
y_test_inv = scaler.inverse_transform(y_test.reshape(-1,1)).flatten()
y_pred_test_inv = scaler.inverse_transform(model.predict(X_test, verbose=0)).flatten()
mape_lstm = mean_absolute_percentage_error(y_test_inv, y_pred_test_inv)
print(f"LSTM 28â€‘day MAPE: {mape_lstm:.2%}")
```

```python
from prophet import Prophet
from sklearn.metrics import mean_absolute_percentage_error

# --- Step 1: Train/Test split ---
daily_qty = daily_qty.reset_index()  # Prophet needs 'ds' column, not index
cutoff_date = daily_qty['ds'].max() - pd.Timedelta(days=28)
train = daily_qty[daily_qty['ds'] <= cutoff_date]
test = daily_qty[daily_qty['ds'] > cutoff_date]

# --- Step 2: Train Prophet model ---
model = Prophet()
model.fit(train)

# --- Step 3: Predict for full range (to cover test period) ---
future = model.make_future_dataframe(periods=28)
forecast = model.predict(future)

# --- Step 4: Evaluate on test period ---
merged = forecast[['ds', 'yhat']].merge(test, on='ds')
mape = mean_absolute_percentage_error(merged['y'], merged['yhat'])
print(f"Prophet 28-day MAPE: {mape:.2%}")

# --- Step 5: Forecast Next 7 Days ---
future_7 = model.make_future_dataframe(periods=7)
forecast_7 = model.predict(future_7)
next_week = forecast_7[['ds', 'yhat']].tail(7)
print("\nNext 7-day forecast using Prophet:")
print(next_week)
```

```python
import pandas as pd

# Step 1: Total pizzas sold by type (based on 'quantity')
pizza_sales = pizza.groupby('pizza_name_id')['quantity'].sum().reset_index()
pizza_sales.rename(columns={'quantity': 'total_sales'}, inplace=True)
```

```python
pizza_sales['percentage'] = pizza_sales['total_sales'] / pizza_sales['total_sales'].sum()
```

```python
total_forecast_qty = next_week['yhat'].sum()  # â‰ˆ 979.22
```

```python
total_forecast_qty
```

```python
pizza_sales['predicted_qty'] = pizza_sales['percentage'] * total_forecast_qty
```

```python
# Merge predicted pizza quantities with ingredients
merged = pizza_sales.merge(ingredients, left_on='pizza_name_id', right_on='pizza_name_id', how='left')

# Calculate how many grams of each ingredient is needed
merged['Ingredient_Required_Grams'] = merged['predicted_qty'] * merged['Items_Qty_In_Grams']
```

```python
merged['Ingredient_Required_Grams']
```

```python
# Sum up by ingredient
final_ingredient_plan = (
    merged.groupby('pizza_ingredients')['Ingredient_Required_Grams']
    .sum()
    .reset_index()
    .sort_values(by='Ingredient_Required_Grams', ascending=False)
)

# Optional: round values
final_ingredient_plan['Ingredient_Required_Grams'] = final_ingredient_plan['Ingredient_Required_Grams'].round(2)
```
